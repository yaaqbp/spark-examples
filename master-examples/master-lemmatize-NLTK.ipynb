```python
import pandas as pd
import nltk
from spacy.lang.en.stop_words import STOP_WORDS
import string
import time
from tqdm import tqdm
tqdm.pandas()
# Download the WordNet data
#nltk.download('wordnet')
```

```python

lemmatizer = nltk.stem.WordNetLemmatizer()
```

```python
start = time.time()
# Define a function to preprocess, lemmatize, and remove stopwords
def preprocess_and_lemmatize(text):
    # Lowercasing
    text = text.lower()
    
    # Remove punctuation
    text = ''.join([char for char in text if char not in string.punctuation])
    
    # Remove extra whitespaces
    lemmatized_text = ' '.join([lemmatizer.lemmatize(word) for word in text.split() if word not in STOP_WORDS])
    
    return lemmatized_text
```

```python
chunks = pd.read_json('data/arxiv-metadata-oai-snapshot.json', lines=True, chunksize = 10000)
```

```python
%%time
chunks = pd.read_json('data/arxiv-metadata-oai-snapshot.json', lines=True, chunksize = 10000)
for idx, df in enumerate(chunks):
    df = df[['id', 'categories','abstract']]
    df = df[df['categories'].str.contains('^\S+$')] #select rows with single cagtegory
    df['processed_text'] = df['abstract'].progress_apply(preprocess_and_lemmatize)
    df.to_csv(f'data/arxiv-lemmatized_preprocessed_NLTK_{idx+1}.csv', index = False)
end = time.time()
```
